```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('full_data.csv')
```

```python
cpi = [296.2, 304.6, 316.3, 315.2, 320.4, 330.5, 337.5, 342.5, 348.3, 348.9, 353.4, 361.0, 369.8]
cpi_adj = [x/369.8 for x in cpi]
```

```python
years = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
infl_dict = dict(zip(years, cpi_adj))

data['real_gprice']=0
for k in infl_dict:
    data.loc[data.year == k, 'real_gprice'] = data['gprice'].map(lambda x: x*infl_dict[k])
    
data.head()
```

```python
data.to_csv('full_data_inflation.csv', index=False)
```
