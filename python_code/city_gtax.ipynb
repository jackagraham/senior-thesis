```python
import pandas as pd
import numpy as np
```

```python
main_df = pd.read_csv('mode_data.csv')
gtax_df = pd.read_csv('state_gtax.csv')
gtax_df = gtax_df.dropna(how='any')
gtax_df = gtax_df.set_index('STATE')
```

```python
main_df.head()
```

```python
x = gtax_df.columns[8:]
gtax_df = gtax_df[x]
gtax_df.head()
```

```python
col = [x for x in main_df.columns]
col.append('tax')

full_df = pd.DataFrame(columns = col)
```

```python
state_map = {'Massachusetts':'Boston', 'Illinois':'Chicago', 'Ohio':'Cleveland'
             ,'Colorado':'Denver', 'Texas':'Houston', 'California':'LA', 'Florida':'Miami', 'New York':'NYC'
             ,'Cali': 'San Francisco', 'Washington':'Seattle' }


def merge_gtaxes(state):
    cit = state_map[state]
    mser = gtax_df.loc[state]
    sdf = pd.DataFrame(mser)
    sdf.columns = ['tax']
    sdf['year'] = sdf.index
    sdf['year'] = [int(x) for x in sdf['year']]
    sdf = sdf.reset_index(drop=True)
    sdf['city'] = cit
    cit_df = main_df[main_df.city==cit]
    cit_df = cit_df.merge(sdf, on=['city','year'])    
    
    return cit_df
```

```python
for state in state_map:
    temp_df = merge_gtaxes(state)
    full_df = pd.concat([full_df, temp_df])
```

```python
full_df.to_csv('full_city_gas_taxes.csv', index=False)
```
